```python
import numpy as np

def feed_forward(inputs, outputs, weights):
    pre_hidden = np.dot(inputs, weights[0]) + weights[1]
    # Apply sigmoid
    hidden = 1 / ( 1 + np.exp(-pre_hidden))
    pred_out = np.dot(hidden, weights[2]) + weights[3]

    mse = np.mean(np.square(pred_out - outputs))
    return mse
```

```python
# Activation Functions

# Tanh
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

def relu(x):
    return np.where(x > 0, x, 0)

def linear(x):
    return x

def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))
```

```python
# Plot examples of activation functions

import matplotlib.pyplot as plt

for i in range(-100, 100):
    x = i / 100
    print(x)
    y_tanh = tanh(x)
    y_relu = relu(x)
    y_linear = linear(x)
    y_softmax = softmax(x)

    # Show in subplots

    plt.subplot(2, 2, 1)
    plt.plot(x, y_tanh, 'bo')
    plt.title('Tanh')

    plt.subplot(2, 2, 2)
    plt.plot(x, y_relu, 'ro')
    plt.title('ReLU')

    plt.subplot(2, 2, 3)
    plt.plot(x, y_linear, 'go')
    plt.title('Linear')

    plt.subplot(2, 2, 4)
    plt.plot(x, y_softmax, 'yo')
    plt.title('Softmax')


plt.show()
```
